## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

This work supports a content team's decision about which pages to prioritize for review and refresh. The unit of analysis is a single page.

Decision: which pages should go into this week's refresh queue, out of thousands of candidates.

Who acts on it: a content strategist or SEO analyst, who pulls the top-ranked pages from the model's output and manually reviews or updates them.

Action: the team spends limited review hours on the pages the model flags as highest priority, rather than reviewing pages at random or by gut feel.

Cost of a wrong call: this cuts both ways. A false negative (a genuinely stale, underperforming page the model ranks low) means it gets skipped, and the business keeps losing traffic or clicks it could have recovered. A false positive (a fine page flagged as high priority) wastes scarce content-team hours on a page that didn't need attention. Since review time is the limited resource, false positives waste effort directly, while false negatives cost a harder-to-see opportunity, so both matter but in different ways.
